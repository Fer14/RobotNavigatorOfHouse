```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
def load_images_from_folder_BN(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv.imread(os.path.join(folder,filename), cv.IMREAD_GRAYSCALE)
        if img is not None:
            images.append(img)
    return images
```

```python
import numpy
def data_set_de_histogramas_desde_imagenes(images,tag):
    DS = []
    for image in images:
        histograma = numpy.concatenate((cv.calcHist([image],[0],None,[256],[0,256]),
                   cv.calcHist([image],[1],None,[256],[0,256]),
                   cv.calcHist([image],[2],None,[256],[0,256])))
        DS.append([histograma/numero_pixeles,
                   tag])
    return DS
```

```python
import numpy
def data_set_de_histogramas_desde_imagenes_BN(images,tag):
    DS_BN = []
    for image in images:
        histograma = (cv.calcHist([image],[0],None,[256],[0,256]))
        DS_BN.append([histograma/numero_pixeles, tag])
    return DS_BN
```

```python
import numpy

# Para ordenar por las distancias
def myFunc(e):
  return e[0]

# Si el mÃ¡s cercano dado un indice es correcto
def leave_one_out(DS, indice):
    comprobando = DS[indice][0]
    landmark_correcto = DS[indice][1]
    distances = []
    for i in range(len(DS)):
        if i != indice:
            distances.append([numpy.linalg.norm(comprobando-DS[i][0]), DS[i][1]])
    distances.sort(key=myFunc)
    return distances[0][1] == landmark_correcto
```

```python
errores_estaticos = 0
for i in range(len(DS)):
    if not leave_one_out(DS, i):
        errores_estaticos += 1
error_estatico = errores_estaticos/len(DS)*100
print('Error estÃ¡tico color: %f' % error_estatico, '%')
```

```python
errores_estaticos_BN = 0
for i in range(len(DS_BN)):
    if not leave_one_out(DS_BN, i):
        errores_estaticos_BN += 1
error_estatico_BN = errores_estaticos_BN/len(DS_BN)*100
print('Error estÃ¡tico blanco y negro: %f' % error_estatico_BN, '%')
```

```python
def one_nn(comprobando, DS):
    distances = []
    for case in DS:
        distances.append([numpy.linalg.norm(comprobando-case[0]), case[1]])
    distances.sort(key=myFunc)
    return distances[0]
```

```python
# Etiquetas correctas
correct = []
for i in range(920):
    if i < 194:
        correct.append(0)
    elif i < 263:
        correct.append(None)
    elif i < 415:
        correct.append(1)
    elif i < 535:
        correct.append(2)
    elif i < 585:
        correct.append(None)
    elif i < 671:
        correct.append(3)
    elif i < 810:
        correct.append(4)
    else:
        correct.append(5)
```

```python
import cv2
vidcap = cv2.VideoCapture('Dinamico_nuevo.mp4')
video_dinamico = cv2.VideoWriter('reconocedor_dinamico.avi', 0, 30, (1280,720))
success,image = vidcap.read()
video = []
count = 0
while success and count < 920:
    hist_to_compare =  numpy.concatenate((cv.calcHist([image],[0],None,[256],[0,256]),
                   cv.calcHist([image],[1],None,[256],[0,256]),
                   cv.calcHist([image],[2],None,[256],[0,256])))/numero_pixeles
    neighbor = one_nn(hist_to_compare, DS)
    # Si la distancia entre imagen y ejemplo mÃ¡s cercano es mayor que 0.09 se clasifica como None
    if neighbor[0] < 0.09:
        video.append(neighbor[1])
    else:
        video.append(None)
    
    # Formato de los textos del vÃ­deo de salida
    image_text_p = 'Predicted: None ' if video[count] is None else 'Predicted: %d ' % video[count]
    image_text_r = '/ Real: None ' if correct[count] is None else '/ Real: %d ' % correct[count]
    image_text = image_text_p + image_text_r
    color = (0,0,0,0)
    if correct[count] == video[count]:
        color = (0, 255, 0, 0)
    else:
        color = (0, 0, 255, 0)
        
    cv2.putText(
         image, #numpy array on which text is written
         image_text, #text
         (10,50), #position at which writing has to start
         cv2.FONT_HERSHEY_SIMPLEX, #font family
         1, #font size
         color, #font color
         3) #font stroke
    video_dinamico.write(image)  
    success,image = vidcap.read()
    count += 1

cv2.destroyAllWindows()
video_dinamico.release()
```
